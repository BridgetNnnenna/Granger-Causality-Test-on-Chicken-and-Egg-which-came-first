```python
from statsmodels.tsa.stattools import kpss, adfuller,grangercausalitytests
import numpy as np
```

```python
#Estimate the number of differencing required to make a time series stationary.
serieseggs = df3['egg in million']
serieschicken = df3['chicken']
```

```python
#Test for Stationarity using ADF test for egg in million
result = adfuller(df3['egg in million'])
result = adfuller(df3['egg in million'].dropna())
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
print()
```

```python
#Interpretation
if result[1] > 0.05:
    print("egg in million is not stationary")
else:
    print("egg in million is stationary")
```

```python
#Test for Stationarity using ADF test for chicken
result = adfuller(df3['chicken'].dropna())
result = adfuller(df3['chicken'])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
print()
```

```python
#Interpretation
if result[1] > 0.05:
    print("chicken is not stationary")
else:
    print("chicken is stationary")
```

```python
#Since Egg in million and Chicken is stationary now then test for causality using granger causality test.

from statsmodels.tsa.stattools import grangercausalitytests
max_lags = 4
#Do Eggs granger cause chickens?
results = grangercausalitytests(df3[['chicken_diff', 'egg in million_diff']], max_lags, verbose=True)
```

```python
lag_to_check = 4
significant_level = 0.05
p_value_lag_4 = results[lag_to_check][0]['ssr_ftest'][1]
```

```python
significant_level = 0.05
if p_value_lag_4 < significant_level:
    print(f"At lag {lag_to_check}, the p-value ({p_value_lag_4}) is below the significant level of {significant_level}, indicating statistical significance and that egg came first")
else:
    print(f"At lag {lag_to_check}, the p-value ({p_value_lag_4}) is above the significant level of {significant_level}, indicating no statistical significance that means eggs did not come first.")
```

```python
# Second order differencing for chicken
df['chicken_diff_2'] = df['chicken_diff'].diff().dropna()
df.head()
```

```python
#Lets drop the NaN values in the dataset
df = df.dropna(subset=['chicken_diff_2'])

df.reset_index(drop=True, inplace=True)

aligned_years = df[df['Year'].isin(df.index)]
df.head()
#Test for Stationarity on the Second differenced data on chicken using ADF test since it wasnt stationary in the previous test
result_diff2_2 = adfuller(df['chicken_diff_2'].dropna())
result_diff2_2 = adfuller(df['chicken_diff_2'])
print('ADF Statistic: %f' % result_diff2_2[0])
print('p-value: %f' % result_diff2_2[1])
print('Critical Values:')
for key, value in result_diff2_2[4].items():
    print('\t%s: %.3f' % (key, value))
print()

if result_diff2_2[1] > 0.05:
    print("chicken_diff_2 is not stationary")
else:
    print("chicken_diff_2 is stationary")
```
